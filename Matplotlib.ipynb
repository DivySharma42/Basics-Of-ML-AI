### Bar charts

```python
df = pd.read_csv('/content/drive/MyDrive/datasets-session-23/batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'], df['2015'])
plt.show()
```

```python
plt.bar(np.arange(df.shape[0])-0.3, df['2015'], width= 0.35)
plt.bar(np.arange(df.shape[0]), df['2016'], width= 0.35)
plt.bar(np.arange(df.shape[0])+0.3, df['2017'], width= 0.35)
plt.xticks(np.arange(df.shape[0]), df['batsman'],rotation = 'vertical')
plt.show()
```

```python
plt.barh(np.arange(df.shape[0])-0.3, df['2015'], )
plt.barh(np.arange(df.shape[0]), df['2016'], )
plt.barh(np.arange(df.shape[0])+0.3, df['2017'], )

plt.show()
```

```python
# stacked bar chart
df
```

```python
plt.bar(df['batsman'], df['2015'])
plt.bar(df['batsman'], df['2016'], bottom=df['2015'])
plt.bar(df['batsman'], df['2017'], bottom=df['2015'])

plt.show()
```
